# 03 — Paper Figures

Displays and annotates each figure produced for the paper, using the pre-saved z = 0.5 artifacts.

```python
import json
from pathlib import Path

import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

ROOT    = Path('/n/home03/hbrittain/COSMO')
FIT_DIR = ROOT / 'artifacts' / 'fit' / 'z0.500'

summary = json.loads((FIT_DIR / 'eft_fit_summary.json').read_text())
scan    = pd.read_csv(FIT_DIR / 'kmax_scan.csv')
curve   = pd.read_csv(FIT_DIR / 'eft_fit_curve.csv')

dof = summary['dof']
print(f"c = {summary['c_counterterm_Mpch2']:.4f} +/- {summary['sigma_c_counterterm_Mpch2']:.4f} (Mpc/h)^2")
print(f"k_stab = {summary['selected_k_stab_h_per_Mpc']:.2f} h/Mpc   ({summary['selection_status']})")
print(f"n_points = {summary['n_points']}   dof = {dof}")
print(f"chi2_lin/dof  = {summary['chi2_linear']/dof:.1f}")
print(f"chi2_SPT/dof  = {summary['chi2_spt']/dof:.1f}")
print(f"chi2_EFT/dof  = {summary['chi2_dof']:.3f}")
print(f"Delta chi2 (SPT->EFT) = {summary['chi2_improvement']:.1f}")
print(f"drift = {summary['parameter_drift_sigma']:.3f} sigma")
print(f"chi2_holdout/dof = {summary['chi2_holdout_dof']:.3f}")
```

### Power Spectrum Comparison

The top panel shows $P(k)$ for the Abacus measurement, the linear baseline $P_{11}$, the one-loop SPT prediction $P_{1\rm loop} = P_{11} + P_{22} + P_{13}$, and the EFT model $P_{\rm EFT} = P_{1\rm loop} - 2c\,k^2 P_{11}$ at $z = 0.5$. All four agree on large scales; differences emerge above $k \sim 0.05\,h/{\rm Mpc}$ where loop contributions become significant.

The bottom panel shows fractional residuals $(P_{\rm model} - P_{\rm sim})/P_{\rm sim}$ on a symlog scale. Linear theory undershoots by ~20% at $k = 0.2\,h/{\rm Mpc}$. One-loop SPT overcorrects — it adds too much power on mildly nonlinear scales. The EFT counterterm ($c = 0.652 \pm 0.028\;({\rm Mpc}/h)^2$) subtracts the UV-sensitive excess and brings the residual to the sub-percent level across the fit range. Fit quality: $\chi^2/{\rm dof} = 1.103$ over 125 data points at $k_{\rm max} = 0.20\,h/{\rm Mpc}$.

```python
fig, ax = plt.subplots(figsize=(9, 7))
ax.imshow(mpimg.imread(FIT_DIR / 'paper_comparison.png'))
ax.axis('off')
fig.tight_layout()
plt.show()
```

### Loop Decomposition

$P_{22}$, $P_{13}$, and their sum $P_{1\rm loop} - P_{11}$ shown separately on a symlog scale. $P_{22}$ is positive at all $k$ — it is a squared integral. $P_{13}$ is negative at the scales we fit and grows in magnitude roughly as $k^2$, partially cancelling $P_{22}$. Their sum is therefore a delicate partial cancellation: by $k = 0.2\,h/{\rm Mpc}$ the loop contribution is ~5% of $P_{11}$, and the counterterm must correct the remaining signed excess that SPT cannot capture.

```python
fig, ax = plt.subplots(figsize=(7, 4))
ax.imshow(mpimg.imread(FIT_DIR / 'loop_decomposition.png'))
ax.axis('off')
fig.tight_layout()
plt.show()
```

### k_max Stability Scan

The top panel traces the best-fit $c$ with $1\sigma$ error bars as the fit range is extended from $k_{\rm max} = 0.10$ to $0.20\,h/{\rm Mpc}$. A genuine EFT counterterm should plateau — it encodes a physical property of the short-mode dynamics (the effective sound speed), so adding data should not change the answer.

The bottom panel shows drift in units of $\sigma_c(k_{\rm ref} = 0.10)$: $\Delta_{\rm drift} = [c(k_{\rm max}) - c(k_{\rm ref})]/\sigma_c(k_{\rm ref})$. The dashed lines mark the $\pm 1\sigma$ acceptance boundary. The maximum drift across all scan points is $0.577\sigma$ — well within the criterion — confirming that $k_{\rm stab} = 0.20\,h/{\rm Mpc}$ is the largest passing scale.

```python
fig, ax = plt.subplots(figsize=(7.5, 7))
ax.imshow(mpimg.imread(FIT_DIR / 'stability_scan.png'))
ax.axis('off')
fig.tight_layout()
plt.show()
```

### Chi-squared vs k_max

$\chi^2/{\rm dof}$ for all three theory levels as a function of $k_{\rm max}$. Linear theory fails by $k \sim 0.1\,h/{\rm Mpc}$ ($\chi^2_{\rm lin}/{\rm dof} = 19.7$ at $k_{\rm max} = 0.20$). One-loop SPT improves things substantially ($\chi^2_{\rm SPT}/{\rm dof} = 5.3$) but still sits far above the dashed line at 1 — the loops overshoot the measured power. The EFT counterterm then absorbs 525 units of $\chi^2$ with a single parameter, bringing the fit to $\chi^2/{\rm dof} = 1.103$.

```python
fig, ax = plt.subplots(figsize=(7.5, 4.5))
ax.imshow(mpimg.imread(FIT_DIR / 'cumulative_chi2.png'))
ax.axis('off')
fig.tight_layout()
plt.show()
```

### Holdout Test

At each $k_{\rm max}$ in the scan we withhold the data in the window $(k_{\rm max},\, k_{\rm max} + 0.04]$, fit $c$ on the remaining range, and evaluate $\chi^2_{\rm holdout}$ on the held-out window. A good model should extrapolate: $\chi^2_{\rm holdout}/{\rm dof} \sim 1$ means the counterterm trained at lower $k$ correctly predicts the data just outside the fit range.

The holdout $\chi^2/{\rm dof}$ at $k_{\rm stab} = 0.20$ is $1.687$ — modestly above 1 but not alarming. This supports the interpretation that the EFT correction is physically grounded rather than a fitting artefact.

```python
fig, ax = plt.subplots(figsize=(7.5, 4.5))
ax.imshow(mpimg.imread(FIT_DIR / 'holdout_scan.png'))
ax.axis('off')
fig.tight_layout()
plt.show()
```
